```python
def maddest(d, axis=None):
    """
    Mean Absolute Deviation
    """
    return np.mean(np.absolute(d - np.mean(d, axis)), axis)

def denoise_signal( x, wavelet='db4', level=1):
    """
    1. Adapted from waveletSmooth function found here:
    http://connor-johnson.com/2016/01/24/using-pywavelets-to-remove-high-frequency-noise/
    2. Threshold equation and using hard mode in threshold as mentioned
    in section '3.2 denoising based on optimized singular values' from paper by Tomas Vantuch:
    http://dspace.vsb.cz/bitstream/handle/10084/133114/VAN431_FEI_P1807_1801V001_2018.pdf
    """
    
    # Decompose to get the wavelet coefficients
    coeff = pywt.wavedec( x, wavelet, mode="per" )
    
    # Calculate sigma for threshold as defined in http://dspace.vsb.cz/bitstream/handle/10084/133114/VAN431_FEI_P1807_1801V001_2018.pdf
    # As noted by @harshit92 MAD referred to in the paper is Mean Absolute Deviation not Median Absolute Deviation
    sigma = (1/0.6745) * maddest( coeff[-level] )

    # Calculte the univeral threshold
    uthresh = sigma * np.sqrt( 2*np.log( len( x ) ) )
    coeff[1:] = ( pywt.threshold( i, value=uthresh, mode='hard' ) for i in coeff[1:] )
    
    # Reconstruct the signal using the thresholded coefficients
    return pywt.waverec( coeff, wavelet, mode='per' )
```

```python
def prepare_list(inlist,toffs=0):
    xlist = []
    for runid in tqdm(inlist):
        dat = pd.read_csv("training/%d.csv" % (runid), header=None)
        dat[0]=dat[0]*1e-6
        dat[0]=np.cumsum(dat[0])
        d1=dat[0]
        d2=dat[1]

        bins=128
        ebins = np.linspace(0,2500,bins+1)
        tmax = int(d1.values[-1])
        X = np.zeros((tmax-toffs,bins))

        for i in range(tmax-toffs):
            dind = np.argwhere((d1 > (toffs + i)) & (d1 < (toffs + i + 1))).flatten() 
            d3 = d2[dind]
            hist = np.histogram(d3, bins=ebins)[0]
            hist = denoise_signal(hist)
            X[i,:] = np.abs(hist)
        xlist.append(X)
    xlist = np.vstack(xlist)
    return xlist

def prepare_siglist(stans, runid_list_filter=None, source_id_arg=None):
    xlist = []
    for row in tqdm(stans.values):
        runid = int(row[0])
        source_id = int(row[1])
        source_time = row[2]
        
        if runid_list_filter is not None and runid not in runid_list_filter:
            continue
        
        if source_id_arg is not None and source_id_arg != source_id:
            continue
        
        dat = pd.read_csv("training/%d.csv" % (runid), header=None)
        dat[0]=dat[0]*1e-6
        dat[0]=np.cumsum(dat[0])
        d1=dat[0]
        d2=dat[1]

        bins=128
        ebins = np.linspace(0,2500,bins+1)
        tmax = int(d1.values[-1])
        X = np.zeros((1,bins))
        source_time = int(source_time-0.5)

        for (i,ti) in enumerate([source_time]):
            dind = np.argwhere((d1 > (ti)) & (d1 < (ti + 1))).flatten() 
            d3 = d2[dind]
            hist = np.histogram(d3, bins=ebins)[0]
            hist = denoise_signal(hist)
            X[i,:] = np.abs(hist)
        xlist.append(X)
    xlist = np.vstack(xlist)
    return xlist
```
