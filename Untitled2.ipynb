```python
import pandas as pd
import pyarrow.parquet as pq
import os
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
%matplotlib inline  
```

```python
trainingAnswers=pd.read_csv("trainingAnswers.csv")
```

```python
tLen = []
for row in tqdm(trainingAnswers.values):
    dat = pd.read_csv("training/%d.csv" % row[0], header=None)
    tLen.append([row[0], np.cumsum(dat[0]).values[-1] * 1e-6, 
                     np.percentile(dat[1], 50),
                     np.percentile(dat[1], 75),
                     np.percentile(dat[1], 90),
                     np.percentile(dat[1], 95),
                     np.percentile(dat[1], 99)])
```

```python
pd.DataFrame(data=tLen,columns=["SourceId","TotTime","P50","P75","P90","P95","P99"]).to_csv("trainTotTime.csv")
```
